```python
### Notebook to Visualize the Indexed Object 
from nexusformat.nexus import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
## Load the object
a=nxload("/data2/2023/chess_mar23/SiP/indexed_objects/SiP_s1_3_15_1_hiRes.nxs")
print(a.tree)
```

```python
### 3D reciprocal space
H=np.array(a.entry.data.H)
K=np.array(a.entry.data.K)
L=np.array(a.entry.data.L)
```

```python
### find H, K, L values in the dimensions of the array 
def fi(H,value):
    for n in range(len(H)):
        if H[n]>value:
            break
    return n-1
```

```python
## plot a slice of reciprocal space
%matplotlib notebook
from matplotlib.colors import LogNorm
import matplotlib
plt.style.use('acs-1col')

plt.imshow(np.rot90(a.entry.data.counts[:,:,fi(L,3)]),extent=[H[0],H[-1],K[0],K[-1]],  norm=LogNorm())
cbar = plt.colorbar(norm = matplotlib.colors.LogNorm())#pcm)
cbar.set_label("Intensity")
plt.xlabel('H (r.l.u.)')
plt.ylabel('K (r.l.u.)')
plt.title("L=3 (r.l.u.)")

plt.tight_layout()
```

```python
### convert array to floats
data_set = np.array(a.entry.data.counts[:,:,:].astype('float32'))

```

```python
### show the linescans
%matplotlib notebook
plt.semilogy(L, data_set[fi(H,-2),fi(K,-2),:], color='blue', label='22')
plt.semilogy(L, data_set[fi(H,-1),fi(K,-1),:], color='red', label = '11')
plt.xlabel("L(r.l.u.)")
plt.ylabel("intensity (log)")
plt.legend()
plt.tight_layout()
```

```python
### Linescans along an H, K column
%matplotlib notebook

im_m1m1 = np.sum(np.sum(data_set[fi(H,-2.05):fi(H,-1.95),fi(K,-2.05):fi(K,-1.95),:], axis=0), axis=0)
im_11 = np.sum(np.sum(data_set[fi(H,-1.05):fi(H,-.95),fi(K,-1.05):fi(K,-.95),:], axis=0), axis=0)

plt.semilogy(L, im_m1m1, label='RT -2-2')
plt.semilogy(L, im_11, label='RT -1-1')
plt.xlabel("L (r.l.u.)")
plt.ylabel("intensity")
plt.legend()
plt.tight_layout()
           
```
